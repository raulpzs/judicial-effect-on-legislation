```python
import pandas as pd

df = pd.read_csv('../data/processed/cases_v2_dejure_lags.csv')

def normalize(col):
    return (
        col.lower()
        .replace('.', '')
        .replace('-', '')
    )

df.columns = [normalize(c) for c in df.columns]
```

```python
lag1_cols = [c for c in df.columns if c.endswith('_lag1')]
lag2_cols = [c for c in df.columns if c.endswith('_lag2')]

base_from_lag1 = {c[:-5] for c in lag1_cols}
base_from_lag2 = {c[:-5] for c in lag2_cols}
```

```python
base_cols = [c for c in df.columns if not c.endswith('_lag1') and not c.endswith('_lag2')]
```

```python
missing_lag1 = [c for c in base_cols if c in base_from_lag1 and f"{c}_lag1" not in df.columns]
len(missing_lag1), missing_lag1[:10]
```

```python
missing_lag2 = [c for c in base_cols if c in base_from_lag2 and f"{c}_lag2" not in df.columns]
len(missing_lag2), missing_lag2[:10]
```

```python
orphan_lag1 = [c for c in lag1_cols if c[:-5] not in df.columns]
orphan_lag2 = [c for c in lag2_cols if c[:-5] not in df.columns]

len(orphan_lag1), len(orphan_lag2)
```

```python
triplets_ok = [
    c for c in base_cols
    if f"{c}_lag1" in df.columns and f"{c}_lag2" in df.columns
]

len(triplets_ok)
```

```python
df.shape
```

```python
new_order = []

for col in df.columns:
    if col.endswith('_lag1') or col.endswith('_lag2'):
        continue

    new_order.append(col)

    if f"{col}_lag1" in df.columns:
        new_order.append(f"{col}_lag1")
    if f"{col}_lag2" in df.columns:
        new_order.append(f"{col}_lag2")

df_reordered = df[new_order]
```

```python
df_reordered.shape
set(df.columns) == set(df_reordered.columns)
```

```python
df_reordered.to_csv('../data/processed/cases_v2_dejure_lags_reordered.csv', index=False)
```
