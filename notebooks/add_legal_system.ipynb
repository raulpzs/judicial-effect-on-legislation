```python
import pandas as pd

pd.set_option("display.max_columns", 200)
```

```python
df = pd.read_csv("../data/processed/judicial_decisions_merged_lags_dejure.csv")

print(df.shape)
df.head()
```

```python
ls = pd.read_csv("../data/raw/ls.csv")

print(ls.shape)
ls.head()
```

```python
print(ls.columns)
print(df.columns)
```

```python
df["country"] = df["country"].astype(str).str.strip()
ls["country"] = ls["country"].astype(str).str.strip()
```

```python
df = df.merge(
    ls[["country", "legal_system"]],
    on="country",
    how="left",
    validate="many_to_one"
)
```

```python
print(df.shape)
print("Missing legal_system:", df["legal_system"].isna().sum())
print(df["legal_system"].value_counts(dropna=False).head(20))
```

```python
df.loc[df["legal_system"].isna(), "country"].drop_duplicates().tolist()
```

```python
df.loc[df["country"] == "European Union", "legal_system"] = "Mixed"
```

```python
print("Missing legal_system:", df["legal_system"].isna().sum())
print(df["legal_system"].value_counts(dropna=False))
```

```python
output_path = "../data/processed/cases.csv"
df.to_csv(output_path, index=False)
print("Saved to:", output_path)
```
